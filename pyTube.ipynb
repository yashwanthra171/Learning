### Pytube can be used to for downloading videos from youtube AND getting metaData about the video

```python
from pytube import YouTube 
from bs4 import BeautifulSoup as bs
from lxml import etree, html

def beautify_html(text):
    doc_root = html.fromstring(text)
    prettyHTML = etree.tostring(doc_root, encoding = 'unicode', pretty_print = True).strip()
    pretty = prettyHTML.split('\n')
    return pretty

def getting_metadata(link, yt, tag):
    metadata = {'watch_html': beautify_html(yt.watch_html), 'embed_html': beautify_html(yt.embed_html)\
                , 'age_restricted':yt.age_restricted, 'js_url': yt.js_url, \
                'js':yt.js, 'initial_data' : yt.initial_data, 'streaming_data': yt.streaming_data,\
                'fmt_streams' : yt.fmt_streams, 'check_availability':yt.check_availability, 'vid_info':yt.vid_info,\
                'bypass_age_gate':yt.bypass_age_gate, 'caption_tracks':yt.caption_tracks, 'captions':yt.captions,\
                'streams':yt.streams, 'thumbnail_url': yt.thumbnail_url,'publish_date':yt.publish_date,\
                'title': yt.title, 'description': yt.description, 'rating':yt.rating, 'length': str(yt.length)+' seconds',\
                'views':yt.views, 'author':yt.author, 'keywords':yt.keywords, 'channel_id': yt.channel_id,\
                'channel_url':yt.channel_url, 'metadata':yt.metadata,\
                'register_on_progress_callback': yt.register_on_progress_callback,\
                'register_on_complete_callback': yt.register_on_complete_callback
               }
    if tag in metadata.keys():
        return metadata[tag]
    else:
        return 'Choose one in :: ', list(metadata.keys())

link = 'https://www.youtube.com/watch?v=xyUxkUPe2xg'
yt = YouTube(link)
```

```python
# js url :: is the vanilla js url parser , that is used to  parse, construct, normalize, and encode urls
# js :: is the text content from the js_url's url output
getting_metadata(link, yt, 'streaming_data')
```
